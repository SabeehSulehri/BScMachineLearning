Generating a random set of two-dimensional points and plotting them as a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0, 50, 20)
y = np.random.randint(0, 50, 20)

plt.scatter(x, y)
plt.show()
```

Plotting the density of the 2D Gaussian distribution N (μ, Σ), with μ=([0,1]) and σ=([1,0.25],[0.25,1]).

```python
from scipy.stats import kde

# Evaluate a gaussian kde on a regular grid of nbins x nbins over
nbins=50
k = kde.gaussian_kde([x,y])
xi, yi = np.mgrid[min(x):max(x):nbins*1j, min(y):max(y):nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))

# Add color bar
plt.pcolormesh(xi, yi, zi.reshape(xi.shape), cmap=plt.cm.jet)
plt.colorbar()
plt.show() 
```
